```python
def fig_insert(fig_loc,fig_caption,fig_label,w):
    S = "\\begin{figure}[h]\n\\begin{center}\n\\includegraphics[width=%f\linewidth]{%s}\n\\caption{%s}\n\\label{%s}\n\\end{center}\n\\end{figure}\n" % (w,fig_loc,fig_caption,fig_label,)
    return S

def fig_rotate(fig_loc,fig_caption,fig_label,w):
    S = "\\begin{sidewaysfigure}[h]\n\\includegraphics[width=%f\linewidth]{%s}\n\\caption{%s}\n\\label{%s}\n\\end{sidewaysfigure}\n" % (w,fig_loc,fig_caption,fig_label,)
    return S
```
